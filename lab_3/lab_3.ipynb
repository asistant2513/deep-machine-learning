```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
fsize=10
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Times New Roman'
plt.rcParams['axes.titlesize'] = fsize
plt.rcParams['axes.labelsize'] = fsize
plt.rcParams['xtick.labelsize'] = fsize
plt.rcParams['ytick.labelsize'] = fsize
plt.rcParams['legend.fontsize'] = fsize
```

```python
x = np.linspace (0.0, 10.0, 100)
fig=plt.figure()
```

```python
ax1=fig.add_subplot(2,2,1)
ax1.plot(x, np.sin(x), 'k-')
ax1.set_xlabel(r'$x$')
ax1.set_ylabel(r'$sin(x)$')
ax1.set_title("A")
```

```python
ax2=fig.add_subplot(2,2,2)
ax2.plot(x, np.cos(x), 'k--', linewidth=1, markersize=3.0)
ax2.set_xlabel(r'$x$')
ax2.set_ylabel(r'$cos(x)$')
ax2.set_title("B")
```

```python
ax3=fig.add_subplot(2,2,3)
ax3.plot(x, (np.sin(x))**2.0, 'k-.', linewidth=1)
ax3.set_xlabel(r'$x$')
ax3.set_ylabel(r'$sin(x)^2$')
ax3.set_title("C")
```

```python
ax4=fig.add_subplot(2,2,4)
ax4.plot(x, (x)**0.15, '-', color='grey', linewidth=1)
ax4.set_xlabel(r'$x$')
ax4.set_ylabel(r'$x^{0.15}$')
ax4.set_title("D")
```

```python
for ax in (ax1, ax2, ax3, ax4):
    ax.set_xlim(0, 2 * math.pi)
    ax.set_ylim(-1.5, 1.5)
```

```python
fig.tight_layout()
fig.savefig("figure.png", orientation='landscape', dpi=300)
```
