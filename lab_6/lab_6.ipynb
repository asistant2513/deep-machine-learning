```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
dframe = pd.read_csv('./titanic_train.csv')
dframe
```

```python
dframe = dframe.drop('Cabin', axis=1)
dframe
```

```python
dframe = dframe.drop(dframe[dframe.isnull().any(axis=1)].index)
dframe
```

```python
sb.pairplot(dframe[['Age', 'Fare', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked', 'Survived']])
```

```python
sb.boxplot(y='Pclass', x='Fare', data=dframe, orient='h')
plt.show()
```

```python
sb.countplot(data=dframe, x='Sex', hue='Survived')
plt.show()
```

```python
sb.countplot(data=dframe, x='Pclass', hue='Survived')
plt.show()
```

```python
dframe['AgePeriod'] = dframe['Age'].map(lambda age: 'Young' if (age < 30) else ('Old' if (age > 60)  else 'Aged'))
```

```python
sb.countplot(data=dframe, x='AgePeriod', hue='Survived')
plt.show()
```
